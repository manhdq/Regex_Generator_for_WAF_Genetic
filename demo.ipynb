```python
requests_list = []

with open('access.log') as f:
    for json_obj in f:
        request = json.loads(json_obj)
        requests_list.append(request)
```

```python
num_accept = 0
num_error = 0

for req in requests_list:
    if int(req['status'].strip()) >= 400:
        num_error += 1
    else:
        num_accept += 1

print(f'Num accept: {num_accept}, Num error: {num_error}')
```
